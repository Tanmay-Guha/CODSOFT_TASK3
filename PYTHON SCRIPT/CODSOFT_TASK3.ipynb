```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/IRIS.csv')

df
```

```python
#Finding the data types of variables in the DataFrame
df.dtypes
```

```python
#Importing libraries essential for data visualization
#MATPLOTLIB
import matplotlib.pyplot as plt
%matplotlib inline

#SEABORN
import seaborn as sns
```

```python
#Plots for continuous variables' analysis
#ENUMERATIVE PLOTS
#UNIVARIATE SCATTER PLOT
plt.scatter(df.index,df['sepal_width'])
plt.show()
```

```python
sns.scatterplot(x=df.index,y=df['sepal_width'],hue=df['species'])
```

```python
#LINE PLOT WITH MARKERS
#Setting title, figure size, labels and font size in matplotlib

plt.figure(figsize=(6,6))
plt.title('Line plot of petal length')
plt.xlabel('index',fontsize=20)
plt.ylabel('petal length',fontsize=20)
plt.plot(df.index,df['petal_length'],markevery=1,marker='d')
```

```python
for name, group in df.groupby('species'):
    plt.plot(group.index, group['petal_length'], label=name,markevery=1,marker='d')
plt.legend()
plt.show()
```

```python
#Setting title, figure size,labels and font size in seaborn
sns.set(rc={'figure.figsize':(7,7)})
sns.set(font_scale=1.5)
fig=sns.lineplot(x=df.index,y=df['petal_length'],markevery=1,marker='d',data=df,hue=df['species'])
fig.set(xlabel='index')
```

```python
#STRIP PLOT
sns.stripplot(y=df['sepal_width'])
```

```python
# Strip-plot(category wise)
sns.stripplot(x=df['species'],y=df['sepal_width'])
```

```python
#SWARM PLOT
#Setting figure size
sns.set(rc={'figure.figsize':(5,5)})

#Swarm-plot
sns.swarmplot(x=df['sepal_width'])

#Swarm-plot category wise
sns.swarmplot(x=df['species'],y=df['sepal_width'])
```

```python
#SUMMARY PLOTS
#HISTOGRAM
plt.hist(df['petal_width'])
sns.distplot(df['petal_width'],kde=False,color='black',bins=10)
```

```python
#DENSITY PLOT
plt.figure(figsize=(5,5))
df['petal_length'].plot(kind='density')
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.kdeplot(df['petal_length'],shade=True)
```

```python
#RUG PLOT
fig, ax = plt.subplots()
sns.rugplot(df['sepal_length'])
ax.set_xlim(3,9)
plt.show()
```

```python
from scipy import stats
import numpy as np
kdf=df['sepal_length'].to_numpy()
rdf=np.hstack(kdf)
density = stats.kde.gaussian_kde(rdf)
x = np.arange(3,9,0.1)
plt.plot(x, density(x))
plt.plot(rdf,[0.01]*len(rdf), '|')
```

```python
sns.distplot(df['sepal_length'],rug=True,hist=False)
```

```python
#BOX PLOT
plt.boxplot(df['sepal_width'])

#Removing the column with categorical variables
dfM=df.drop('species',axis=1)
plt.figure(figsize=(9,9))
```

```python
#Set Title
plt.title('Box plots of the 4 variables')
plt.boxplot(dfM.values,labels=['SepalLength','SepalWidth','PetalLength','PetalWidth'])
```

```python
sns.boxplot(df['sepal_width'])
```

```python
sns.set(rc={'figure.figsize':(9,9)})
sns.boxplot(x="variable", y="value", data=pd.melt(dfM))
```

```python
#distplot()
sns.set(rc={'figure.figsize':(6,6)})
sns.distplot(df['petal_length'],color='black',rug=True)
```

```python
#VIOLIN PLOT
plt.figure(figsize=(7,7))
plt.violinplot(dfM.values,showmedians=True)
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.violinplot(df['sepal_width'],orient='vertical')
```

```python
sns.set(rc={'figure.figsize':(9,9)})
sns.violinplot(x=df['species'], y=df['petal_width'],data=df)

```
